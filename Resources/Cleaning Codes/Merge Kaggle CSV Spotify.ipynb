```python
#read in tracks CSV
tracks = pd.read_csv( '../CSV files/tracks.csv.zip')
tracks
```

```python
# split multiple artists/ songs from one line
clean = tracks.assign(id_artists=tracks.id_artists.str.split(",")).explode('id_artists')
clean
```

```python
# split multiple artists/ songs from one line
clean = tracks.assign(artists=tracks.artists.str.split(",")).explode('artists')
clean
```

```python
# build function to remove unwanted square brackets and qoutations

def clean_up_lists(x):
    return x.replace(']','').replace('[','').replace("'","")
    
```

```python
#remove square brackets and qoutations
Artists["genres"] = Artists["genres"].apply(clean_up_lists)
Artists
```

```python
# replace any blank values with NaN and drop 

Artists.replace('', np.nan, inplace = True)
Artists = Artists.dropna()
Artists
```

```python
# Artists.to_csv('genre_artist.csv', index=False)
```

```python
#remove square brackets and qoutations
clean["id_artists"] = clean["id_artists"].apply(clean_up_lists)
clean
```

```python
#remove square brackets and qoutations
clean["artists"] = clean["artists"].apply(clean_up_lists)
clean
```

```python
# rename columns to merge
tracks = clean.rename(columns={'name': 'song', 'artists':'artist'})
tracks
```

```python
# filter tracks dates 2010 & 2020
start_date = "2010-1-1"
end_date = "2020-12-31"

after_start_date = tracks["release_date"] >= start_date
before_end_date = tracks["release_date"] <= end_date
between_two_dates = after_start_date & before_end_date
filtered_dates = tracks.loc[between_two_dates]




# filtered_dates = tracks['release_date'].loc['2010':'2020']
filtered_dates
```

```python
filtered_dates.to_csv('filtered_tracks.csv', index=False)
```

```python
# merge Artist with Tracks to include genre on dataframe
# A_tracks = pd.merge(Artists,filtereddates, how='left',left_on=['A','B'],right_on=['A','CC']).dropna()

A_Tracks = pd.merge(Artists,filtered_dates, how='left', on='popularity')
A_Tracks
```
