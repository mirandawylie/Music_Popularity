```python
music_vids = pd.read_csv( 'kpop_music_videos.csv')
music_vids
```

```python
cleanidols = idols.drop(columns=['Korean Name', 'K. Stage Name', 'Birthplace', 'Other Group'])
```

```python
cleanmusic = music_vids.drop(columns=['Korean Name', 'Director'])
```

```python
merged = pd.merge(cleanidols, cleanmusic, left_on=  ['Group'],
                   right_on= ['Artist'], 
                   how = 'left')
merged
```

```python
order = [7,3,10,0,5,2,1,6,8,9,11,4] # setting column's order
kpop_df = merged[[merged.columns[i] for i in order]]

kpop_df
```
